```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5,6],
    'y': [9,5,7,3,6,1],
    'z': [2,3,4,5,3,2]
})
plt.plot(df['x'], df[['y','z']])
plt.grid(True)
plt.xticks(np.arange(6), list('abcdef'))
plt.yticks(np.arange(0,10,step=1), rotation=45)
plt.title('Tes Chart')
# plt.style.available
plt.show
```

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5,6],
    'y': [9,5,7,3,6,1],
    'z': [2,3,4,5,3,2]
})

plt.figure('Test', figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['x'], df[['y','z']])

plt.subplot(1, 2, 2)
plt.plot(df['x'], df[['y','z']])
plt.grid(True)
plt.xticks(np.arange(6), list('abcdef'))
plt.yticks(np.arange(0,10,step=1), rotation=45)
plt.title('Tes Chart')

plt.suptitle('Ini super title')


#Double figure
plt.figure('Cek', figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['x'], df[['y','z']])

plt.subplot(1, 2, 2)
plt.plot(df['x'], df[['y','z']])
plt.grid(True)
plt.xticks(np.arange(6), list('abcdef'))
plt.yticks(np.arange(0,10,step=1), rotation=45)
plt.title('Tes Chart')

plt.suptitle('Ini super title')
# plt.style.available
plt.show
```

```python
plt.figure('Test', figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['x'], df[['y','z']])

plt.subplot(1, 2, 2)
plt.bar(df['x'], df['y'], color='pink')
plt.grid(True)
plt.xticks(np.arange(7))
plt.yticks(np.arange(0,10,step=1), rotation=45)
plt.title('Tes Chart', fontdict={'family':'arial'})

plt.suptitle('Ini super title')
```

```python
plt.figure('Test', figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['x'], df[['y','z']])

plt.subplot(1, 2, 2)
plt.bar(df['x'], df['y'], color=['m','b','r','g','b', 'm'])
plt.grid(True)
plt.xticks(np.arange(7))
plt.yticks(np.arange(0,10,step=1), rotation=45)
plt.title('Tes Chart', fontdict={'family':'arial'})
for i in range(len(df['x'])):
    plt.text(df['x'].iloc[i], df['y'].iloc[i]+0.1, df['y'].iloc[i])

plt.suptitle('Ini super title')
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(df['x'], df[['y','z']])

plt.subplot(122)
plt.bar(df['x'], df['y'])
plt.bar(df.x, df.z, bottom=df.y)
plt.grid(True)
plt.xticks(np.arange(7))
plt.yticks(np.arange(0,10,step=1), rotation=45)
plt.title('Tes Chart', fontdict={'family':'arial'})

plt.show()
```
